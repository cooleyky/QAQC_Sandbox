```python
deployment_tracking = pd.DataFrame(columns=['Array','Num of Deployments'])
for file in os.listdir('/home/andrew/Documents/OOI-CGSN/asset-management/deployment/'):
    if 'CE' in file or 'RS' in file:
        continue
    # Load the csv into a pandas dataframe
    df = pd.read_csv('/home/andrew/Documents/OOI-CGSN/asset-management/deployment/'+file)
    # Remove the commented out deployments
    df = df[df['CUID_Deploy'].apply(lambda x: False if x.startswith('#') else True)]
    # Count the number of unique deployments based on the unique cruises
    deployments = len(np.unique(df['CUID_Deploy']))
    # Put the results into a new dataframe for easy tracking
    deployment_tracking = deployment_tracking.append({
        'Array':file.split('_')[0],
        'Num of Deployments':deployments
    }, ignore_index=True)
```

```python
deployment_tracking.sort_values(by='Array', inplace=True)
```

```python
deployment_tracking.to_csv('/home/andrew/Documents/OOI-CGSN/QAQC_Sandbox/Metadata_Review/Deploy_stats.csv')
```
